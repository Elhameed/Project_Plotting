```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the histogram with different bin widths
co2_data = df['co2']
print(co2_data)

sns.set_style('whitegrid')
sns.histplot(data = co2_data, bins=30, color='powderblue', alpha=0.7, edgecolor="white")

# Add labels and title
plt.title('Distribution of carbon Dioxide')
plt.xlabel('co2')
plt.ylabel('Frequency')
plt.savefig('./chart_images/c02.png')
plt.show()
```

```python
!mkdir chart_images
```

**Interpretation:**
The distribution of carbon dioxide (CO2) levels is shown by the histogram. The frequency of CO2 values in each bin is indicated by the y-axis, while CO2 concentration ranges are displayed on the x-axis. I also used certain attributes like 'whitegrid' style which displays a white background with grid lines, alpha to display transparency in order to beautify the histogram

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the histogram with different bin widths
UCity_data = df['UCity']
print(UCity_data)

sns.set_style('whitegrid')
sns.histplot(data = UCity_data, bins=30, color='yellow', alpha=0.8, edgecolor="black")

# Add labels and title
plt.title('Histogram of Urban City mpg')
plt.xlabel('UCity')
plt.ylabel('Frequency')
plt.savefig('./chart_images/UCity.png')
plt.show()
```

**Interpretation:**
The distribution of urban city(UCity)mpg values is represented by this histogram. The frequency of UCity values within each bin is shown on the y-axis, and the x-axis shows the UCity values grouped into bins. The color and style of the plot make the distribution pattern more visible. The typical range and frequency of urban fuel efficiency ratings are analyzed with the aid of this histogram.

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the histogram with different bin widths
comb_data = df['comb']
print(comb_data)

sns.set_style('whitegrid')
sns.histplot(data = comb_data, bins=30, color='purple', alpha=0.6, edgecolor="powderblue")

# Add labels and title
plt.title('Histogram of comb')
plt.xlabel('comb')
plt.ylabel('Frequency')
plt.savefig('./chart_images/comb.png')
plt.show()
```

**Interpretation: **
The distribution of combined fuel efficiency (comb) values is represented by this histogram. The frequency of data points in each bin is shown on the y-axis, while the x-axis displays several comb value ranges, or bins.

```python
numeric_values= df.select_dtypes(include=[ 'number'])
print (numeric_values)
corr = numeric_values.corr()
plt.figure(figsize=(8,6))
sns. heatmap(corr, annot=True, cmap="coolwarm", fmt=' .2f')
plt.xticks (rotation=45)
plt. title('Correlation map')
plt.savefig('./chart_images/heatmap.png')
plt. show()
```

**Interpretation:**
The heatmap illustrates the correlation matrix of numeric variables in the dataset. Each cell represents the strength and direction of relationships between variables, ranging from blue which stands for negative correlation to red which stands for positive correlation. Darker shades indicate stronger correlations. This visualization reveals which variables are closely related and which are independent.
